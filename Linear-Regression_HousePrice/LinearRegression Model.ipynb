```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

def dataprocessing1(filepath):
    df = pd.read_csv(filepath)
    numofnull = df.isnull().sum().sum()
    df = df.fillna(df.mean())
    return numofnull, df

filepath = r'C:\Users\HP\Desktop\Housingcustomdata.csv'
numofnull, df = dataprocessing1(filepath)




df['Bedroom/Area'] = df['Bedroom'] / df['Area']


df.to_csv(r'C:\Users\HP\Desktop\Housingcustomdata_with_new_feature.csv', index=False)
df.head()

x = df.drop(columns=['SalePrice'])
y = df['SalePrice']

def splitting(): 
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    return model, X_train, X_test, y_train, y_test

model, X_train, X_test, y_train, y_test = splitting()


y_pred = model.predict(X_test)

plt.figure(figsize=(12, 6))



BA = np.sort(df['Bedroom/Area'].values)
SP = np.sort(df['SalePrice'].values)
plt.plot(BA, SP, color='red', linewidth=2, label='Regression line')
plt.xlabel('Bedroom/Area')
plt.ylabel('SalePrice')
plt.title('Bedroom/Area vs SalePrice')
plt.show()



plt.scatter(y_test, y_pred, color='blue', label='Test vs Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red',  marker='x', linewidth=2, label='Line of Best Fit')
plt.xlabel('Actual SalePrice')
plt.ylabel('Predicted SalePrice')
plt.title('Actual vs Predicted SalePrice')
plt.legend()


plt.show()


y_pred = model.predict(X_test)
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")

def predict_value(model):
    print("Please enter the following values:")
    
   
    features = ['Area', '1st sqft', '2nd sqft', 'FullBath', 'HalfBath', 'Bedroom']
    
    user_data = {}
    for feature in features:
        while True:
            try:
                value = float(input(f"{feature}: "))
                user_data[feature] = value
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")
    

    user_df = pd.DataFrame([user_data], columns=features)
    
  
    predicted_value = model.predict(user_df)
    print(f"Predicted SalePrice: ${predicted_value[0]}")

predict_value(model)
```
